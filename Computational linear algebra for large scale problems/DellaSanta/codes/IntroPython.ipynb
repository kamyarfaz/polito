## If, For, While Blocks

Here we show some (very) simple examples concernig the if-elif-else, for, and while blocks

```python
myrange = range(0, 15, 2)

for n in myrange:
    if n <= 10:
        print(f'{n} is less than 11')
    elif 10 < n <= 12:
        print(f'{n} is between 11 and 12!')
    else:
        print(f'{n} greater than 12!')
```

```python
counter = 0
n = myrange[counter]

while n <= 10:
    print(f'Element {counter + 1} has value {n} (less than 11)')
    counter += 1
    n = myrange[counter]
```
